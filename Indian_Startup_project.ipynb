```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/sample_data/startup_funding.csv')
df
```

```python
# Settings the columns.
df.columns=['SNo', 'Date', 'StartupName', 'IndustryVertical', 'SubVertical',
       'CityLocation', 'InvestorsName', 'InvestmentType', 'AmountInUSD',
       'Remarks']
```

```python
# Let's remove inconsistency.
df['StartupName'] = df['StartupName'].apply(lambda x:(str(x).replace("\\\\"," ")))
df['StartupName'] = df['StartupName'].apply(lambda x:(str(x).replace("\\"," ")))
df['StartupName'] = df['StartupName'].apply(lambda x:(str(x).replace("\""," ")))
```

```python
# To see the percentage of missing value in every column.
m = (100-(df.count()/df.isna().count())*100)
display(m)
```

```python
# As we see, Remark column has the 86% missing value so we will drop it.
df.drop("Remarks", axis=1,  inplace=True) 
```

```python
df.head()
```

```python
missing_values = df.isnull().sum()
missing_value_frame = missing_values.to_frame()

missing_value_frame.columns=['count']
missing_value_frame.index.names = ['Name']
missing_value_frame['Name'] = missing_value_frame.index
```

```python
missing_value_frame.head()
```

```python
# Graphical View of missing value.
plt.figure(figsize=(10,5))
sns.set(style="whitegrid", color_codes = True)
sns.barplot(x = 'Name', y = 'count', data = missing_value_frame);
plt.title("Missing value in dataframe")
plt.xticks(rotation=45);

```
